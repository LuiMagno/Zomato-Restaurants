# Fome Zero - País

## 01 -  Qual o nome do país que possui mais cidades registradas?

```python
# Pegando os valores únicos de cidades por países
cidades_por_pais = df.loc[:, ['country_name', 'city']].groupby('country_name').nunique().sort_values(by='city').reset_index()
print(f'O pais que mais possui cidades registrada é: {cidades_por_pais.iloc[-1,0]}')
cidades_por_pais
```

## 02 - Qual o nome do país que possui mais restaurantes registrados?

```python
# Valores únicos de restaurantes por países
restaurantes_por_pais = df.loc[:, ['country_name', 'restaurant_id']].groupby('country_name').nunique().sort_values(by='restaurant_id').reset_index()
print(f'O pais que mais possui restaurantes registrados é: {restaurantes_por_pais.iloc[-1,0]}')
restaurantes_por_pais
```

## 03 - Qual o nome do país que possui mais restaurantes com o nível de preço igual a 4 registrados?

```python
cols = ['country_name', 'restaurant_id']
restaurantes_nivel4 = df.loc[df['price_range'] == 'gourmet', cols].groupby('country_name').count().sort_values(by='restaurant_id').reset_index()
print(f'O pais que mais possui restaurantes com nível gourmet é: {restaurantes_nivel4.iloc[-1,0]}')
restaurantes_nivel4
```

## 04 - Qual o nome do país que possui a maior quantidade de tipos de culinária distintos?

```python
cols = ['country_name', 'cuisines']
paises_tipos_culinaria = df.loc[:, cols].groupby('country_name').nunique().sort_values(by='cuisines').reset_index()
print(f'O pais que possui a maior quantidade de tipos de culinária é: {paises_tipos_culinaria.iloc[-1,0]}')
paises_tipos_culinaria
```

## 05 -  Qual o nome do país que possui a maior quantidade de avaliações feitas?

```python
avaliacoes_por_pais = df.loc[:, ['country_name', 'votes']].groupby('country_name').sum().sort_values(by='votes').reset_index()
print(f'O pais que possui a maior quantidade de avaliacões é: {avaliacoes_por_pais.iloc[-1,0]}')
avaliacoes_por_pais
```

## 06 - Qual o nome do país que possui a maior quantidade de restaurantes que fazem entrega?

## 07 - Qual o nome do país que possui a maior quantidade de restaurantes que aceitam reservas?

```python
cols = ['country_name', 'has_table_booking']
pais_reservas = df.loc[df['has_table_booking'] == 1, cols].groupby('country_name').count().sort_values(by='has_table_booking').reset_index()
print(f'O pais que possui a maior quantidade de restaurantes que fazem reserva é: {pais_reservas.iloc[-1,0]}')
pais_reservas
```

## 08 - Qual o nome do país que possui, na média, a maior quantidade de avaliações registrada?

## 09 - Qual o nome do país que possui, na média, a maior nota média registrada?
